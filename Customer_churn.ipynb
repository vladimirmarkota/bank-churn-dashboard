```python
import pandas as pd
df = pd.read_csv("Customer-Churn-Records.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
bins = [18, 30, 45, 60, 100]
labels = ['18-30', '31-45', '46-60', '60-100']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
churn_rate = df['Exited'].mean()
print(f"Churn rate: {churn_rate*100:.2f}%")
```

```python
total_clients = len(df)
print(f"Total clients: {total_clients}")
```

```python
avg_balance = df.groupby('Exited')['Balance'].mean()
print(avg_balance)
```

```python
avg_products = df.groupby('Exited')['NumOfProducts'].mean()
print(avg_products)
```

```python
age_dist = df['AgeGroup'].value_counts()
print(age_dist)
```

```python
gender_churn = df.groupby(['Gender','Exited']).size().unstack()
print(gender_churn)
```

```python
begender_churn=df.groupby(['Gender','Exited]).size()
print(begender_churn)
```

```python
geo_dist = df['Geography'].value_counts()
print(geo_dist)
```

```python
df.to_csv("cleaned_bank_data.csv", index=False)
```
